```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#ساعات مذاكرة مقابل درجات
x = np.array([1,2,3,4,5,6]).reshape(-1,1)
y = np.array([1.5,2.0,2.5,3.5,3.7,4.5])
```

```python
#انشاء النموذج و تدريبه
model = LinearRegression()
model.fit(x,y)
```

```python
y_new = model.predict([[7]])
print("القيمة المتوقعة عند x = 7: ",y_new[0])
print("coefficient: ",model.coef_[0])
print("intercept: ",model.intercept_)
```

```python
y_pred = model.predict(x)
print("expected values: ",y_pred)
```

```python
mse = mean_squared_error(y,y_pred)
print("mean squared error: ",mse)
```

```python
plt.scatter(x,y,color = 'green', label = 'original data')
plt.plot(x,y_pred, color = 'red' , label = 'regression line')
plt.xlabel("study hours")
plt.ylabel("grades")
plt.legend()
plt.show()
```
