```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from mpl_toolkits.mplot3d import Axes3D
```

```python
np.random.seed(42)
data_3d = np.random.rand(200,3)
```

```python
k_values = [2,3,4]
result = {"3d":[]}
```

```python
for k in k_values:
    Kmeans_3d = KMeans(n_clusters=k, random_state=42,).fit(data_3d)
    labels = Kmeans_3d.labels_
    score = silhouette_score(data_3d, labels)
    result["3d"].append((k, score))
    fig = plt.figure()
    ax = fig.add_subplot(111, projection="3d")
    ax.scatter(data_3d[:,0],data_3d[:,1], c=labels, cmap="viridis")
    ax.scatter(Kmeans_3d.cluster_centers_[:,0],Kmeans_3d.cluster_centers_[:,1],c="black", marker="o", s=200,label="ceners")
    ax.set_title(f"kmeans 3d - k= {k}, silhouette= {score:.3f}")
    plt.legend()
    plt.show()
```
